```python
# Import Libraries
import requests
import datetime
```

```python
today = datetime.datetime.today()
params = {'date': today.strftime("%Y-%m-%d")} # YYYY-MM-DD
```

```python
res = requests.get('https://api.data.gov.sg/v1/environment/rainfall', params=params)
```

```python
res.status_code
```

```python
rainfall = res.json()
```

```python
rainfall
```

```python
rain = rainfall['metadata']['stations']
```

```python
import pandas as pd
```

```python
pd.DataFrame.from_dict(rain)
```
